### Importação de Arquivos e Bibliotecas



```python
import pandas as pd
import shutil
from pathlib import Path 
import yagmail

Lojas_df = pd.read_csv(r'/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Bases de Dados/Lojas.csv', encoding = 'iso-8859-1', sep=';')

Vendas_df = pd.read_excel(r'/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Bases de Dados/Vendas.xlsx')

Emails_gerentes_df = pd.read_excel(r'/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Bases de Dados/Emails.xlsx')


display(Lojas_df)
display(Vendas_df)
display(Emails_gerentes_df)
```

### Definição e  Criação  de uma planilha para cada Loja, criação de pasta para cada loja e download de cada planilha de cada loja para sua pasta especifica

```python
lista_Lojas = ['IguatemiEsplanada_df','ShoppingMidwayMall_df','NorteShopping_df','ShoppingIguatemiFortaleza_df','ShoppingUniãodeOsasco_df','ShoppingCenterInterlagos_df','RioMarRecife_df','SalvadorShopping_df','RioMarShoppingFortaleza_df','ShoppingCenterLesteAricanduva_df','RibeirãoShopping_df','ShoppingMorumbi_df','ParqueDomPedroShopping_df','BourbonShoppingSP_df','PalladiumShoppingCuritiba_df','PasseidasÁguasShopping_df','CenterShoppingUberlândia_df','ShoppingRecife_df','ShoppingVilaVelha_df','ShoppingSPMarket_df','ShoppingEldorado_df','ShoppingIbirapuera_df','NovoShoppingRibeirãoPreto_df','IguatemiCampinas_df','ShoppingBarra_df']
nome_Lojas = lista_Lojas[:]


for nome in nome_Lojas:
    Path('/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Backup Arquivos Lojas/{}'.format(nome)).mkdir()


for i, loja in enumerate(Lojas_df['Loja']):
    print(loja)
    if loja == Emails_gerentes_df['Loja'][i]:
        lista_Lojas[i] = Vendas_df[Vendas_df['ID Loja'] == i+1]
        local_arquivo = '/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Backup Arquivos Lojas/{}/{}.xlsx'.format(nome_Lojas[i],nome_Lojas[i] )
        lista_Lojas[i].to_excel(local_arquivo)
         
```

###  Calculo de indicadores de meta definidos pela empresa para todas as lojas

```python
Faturamento_Meta_Dia = 1000
Faturamento_Meta_Ano = 1650000


Diversidade_Meta_Dia = 4
Diversidade_Meta_Ano = 120


Ticket_média_Meta_Dia = 500
Ticket_média_Meta_Ano = 500
```

### Envio automatico de email para cada gerente com relatório de sua devida loja

```python
Lojas_dict = {}

for i,loja in enumerate(lista_Lojas):
    Lojas_dict[Lojas_df['Loja'][i]] = loja

for i,loja in enumerate(Lojas_dict):
    vendas_Lojas = Lojas_dict[loja]
    
    ultimos_dias = vendas_Lojas['Data'] == vendas_Lojas['Data'].iloc[-1]
    
    ultimos_dias_Lojas_df = vendas_Lojas[ultimos_dias]
    
    
    
    
    # Faturamento ano
    Faturamento_ano = vendas_Lojas['Valor Final'].sum()
    # Faturamento dia
    Faturamento_dia = vendas_Lojas['Valor Final'].iloc[-1]
    
    # Diversidade de produtos ano
    Diversidade_produto_ano = len(vendas_Lojas['Produto'].unique())
    # Diversidade de produtos dia
    Diversidade_produto_dia = len(ultimos_dias_Lojas_df['Produto'].unique())
    
    
    # Ticket médio ano
    Média_vendas_ano = vendas_Lojas.groupby('Código Venda')[['Valor Final']].sum()
    Ticket_medio_ano = Média_vendas_ano['Valor Final'].mean()
    # Ticket médio 
    Média_vendas_dia = ultimos_dias_Lojas_df.groupby('Código Venda')[['Valor Final']].sum()
    Ticket_medio_dia = ultimos_dias_Lojas_df['Valor Final'].mean()
    
    
    # Cenarios
    
    if Faturamento_dia > Faturamento_Meta_Dia:
        Cenario_Faturamento_dia = "Positivo"
    else:
        Cenario_Faturamento_dia = "Negativo"
        
    if Faturamento_ano > Faturamento_Meta_Ano:
        Cenario_Faturamento_ano = "Positivo"
    else:
        Cenario_Faturamento_ano = "Negativo"
        
        
    
    if Diversidade_produto_dia > Diversidade_Meta_Dia:
        Cenario_Diversidade_dia = "Positivo"
    else:
        Cenario_Diversidade_dia = "Negativo"
        
    if Diversidade_produto_ano > Diversidade_Meta_Ano:
        Cenario_Diversidade_ano = "Positivo"
    else:
        Cenario_Diversidade_ano = "Negativo"
    
    
    
    if Ticket_medio_dia > Ticket_média_Meta_Dia:
        Cenario_Ticket_dia = "Positivo"
    else:
        Cenario_Ticket_dia = "Negativo"
        
    if Ticket_medio_ano > Ticket_média_Meta_Ano:
        Cenario_Ticket_ano = "Positivo"
    else:
        Cenario_Ticket_ano = "Negativo"
    
    # Envio de emails
    email_gerente = Emails_gerentes_df['E-mail'][i]
    gerente = Emails_gerentes_df['Gerente'][i]
    usuario = yagmail.SMTP(user='lucasluisouza@gmail.com',                                                  password='lucasluis2004' )
    usuario.send(to=[email_gerente, 'lucasluisouza@gmail.com'],
    subject='Relatorio da loja {}'.format(loja),
    contents= f''' Ola {gerente}.
    O resultado de ontem {ultimos_dias_Lojas_df['Data'].iloc[-1]}.
    
    
    
    Indicador    |  Valor Dia  |    Meta Dia    |  Cenário Dia  |
    
    Indicador 1  |  R$ {Faturamento_dia:,.2f}  |  R$ {Faturamento_Meta_Dia:>9,.2f}  |    {Cenario_Faturamento_dia}   |
    
    Indicador 2  |  {Diversidade_produto_dia:>5}      |  {Diversidade_Meta_Dia:>5}         |    {Cenario_Diversidade_dia}   |
    
    Indicador 3  |  R$ {Ticket_medio_dia:.2f}  |  R$ {Ticket_média_Meta_Dia:>4}       |    {Cenario_Ticket_dia}   |
    
    
    
    Indicador    |     Valor Ano     |     Meta Ano      |  Cenário Ano  |
    
    Indicador 1  |  R$ {Faturamento_ano:>2,.2f}  |  R$ {Faturamento_Meta_Ano:,.2f}  | {Cenario_Faturamento_ano:>12}  |
    
    Indicador 2  | {Diversidade_produto_ano:>11}       | {Diversidade_Meta_Ano:>11}       | {Cenario_Diversidade_ano:>12}  |
    
    Indicador 3  |  R$ {Ticket_medio_ano:>9.2f}     |  R$ {Ticket_média_Meta_Ano:>9.2f}     | {Cenario_Ticket_ano:>12}  |
    
    
    Segue em anexo a planilha com todos os dados para mais detalhes.
    
    
    Qualquer dúvida estou à disposição.
    
    Att., Lucas Luis
    
    
    ''',
    attachments='/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Backup Arquivos Lojas/{}/{}.xlsx'.format(nome_Lojas[i],nome_Lojas[i] ))
    
```

### Criando ranking de lojas para enviar para a diretoria

```python
Lista_valores = {}

Ranking_dia = pd.DataFrame()

Ranking_dia['Loja'] = Lojas_df['Loja']

Ranking_dia['Valor Final'] = 0


Ranking_ano = pd.DataFrame()

Ranking_ano['Loja'] = Lojas_df['Loja']

Ranking_ano['Valor Final'] = 0


for i,loja in enumerate(Lojas_dict):
    vendas_Lojas = Lojas_dict[loja]
    
    
    ultimos_dias = vendas_Lojas['Data'] == vendas_Lojas['Data'].iloc[-1]
    
    
    ultimos_dias_Lojas_df = vendas_Lojas[ultimos_dias]
    
    
    
    ultimos_dias_Lojas_df = ultimos_dias_Lojas_df.groupby('ID Loja')[['Valor Final']].sum()
    
    
    Lista_valores[loja] = ultimos_dias_Lojas_df['Valor Final'].iloc[-1]
    
    
    Ranking_dia['Valor Final'].loc[i] = Lista_valores[loja]
    
    Ranking_dia = Ranking_dia.sort_values('Valor Final', ascending=False)
    
    
    Ranking_ano['Valor Final'].loc[i] = vendas_Lojas['Valor Final'].sum()
    
    Ranking_ano = Ranking_ano.sort_values('Valor Final', ascending=False)
    
   
    
display(Ranking_dia)
display(Ranking_ano)

Path('/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Diretoria/Rankings').mkdir()

Ranking_ano.to_excel('/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Diretoria/Rankings/Ranking ano.xlsx')


Ranking_dia.to_excel('/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Diretoria/Rankings/Ranking dia.xlsx')



shutil.make_archive('Rankings', 'zip', '/home/lucasluis/Python Impressionador/Projeto de automação de indicadores/Projeto Automacao Indicadores/Diretoria/', 'Rankings')
```
